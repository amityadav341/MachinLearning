```python
import matplotlib.pyplot as plt
x=[2013,2014,2015,2016,2017,2018,2019,2020]
y=[50,47,52,53,55,46,48,53]
plt.scatter(x,y,label='old',color='red',marker='o',s=50)
y=[30,33,28,38,27,31,35,30]
plt.xlabel('years')
plt.ylabel('Tuberculosis')
plt.scatter(x,y,label='young',color='g',marker='o',s=50)
plt.legend()
```

```python
import matplotlib.pyplot as plt
x=[15,20,50,70]
a=['old','young','women','men']
plt.pie(x,labels=a,colors=['r','g','b','m'],startangle=90,autopct="%1.2f%%",explode=(0.1,0,0,0))
plt.legend()
```

```python
import matplotlib.pyplot as plt
import csv
x=[]
y=[]
with open('data.txt','r')as csvfile:
    plots=csv.reader(csvfile)
    for col in plots:
        x.append(col[0])
        y.append(col[1])
plt.xlabel('years')
plt.ylabel('patient')
plt.title('tuberculosis')
plt.plot(x,y,marker='o',color='r')
plt.bar(x,y)
```
